```python
from pykeen.predict import predict_triples
from pykeen.datasets import get_dataset
dataset = get_dataset(dataset="nations")
pack = predict_triples(model=pipeline_result.model, triples=dataset.validation)
```

```python
df = pack.process(factory=pipeline_result.training).df
```

```python
df.nlargest(n=30, columns="score")
```

```python
pipeline_result.training
```

```python
pack_testing = predict_triples(model=pipeline_result.model, triples=dataset.testing)
```

```python
df_testing = pack_testing.process(factory=pipeline_result.training).df
```

```python
df_testing.nlargest(n=20, columns="score")
```

```python
from pykeen.evaluation import LCWAEvaluationLoop
evaluation_loop = LCWAEvaluationLoop(
    model=pipeline_result.model,
    triples_factory=dataset.testing,
)
```

```python
evaluation = evaluation_loop.evaluate()
```
